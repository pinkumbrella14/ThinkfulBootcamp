```python
#upcase everything in journal title column
data_original['j_title'] = data_original['Journal title'].str.upper()
```

```python
#get rid of leading and trailing space in the string journal title column
data_original['j_title'] = data_original['j_title'].str.strip() 
```

```python
#what are all the different journals that we are dealing with?
data_original.groupby('j_title').count()
```

```python
#create a new journal title column and keep it blank for now
data_new = data_original
data_new['new_j_title'] = ''
```

```python
#subset data based on journal titles with word 'academy' in it
data_original[(data_original['j_title'].str.contains('ACADEMY', case = False, regex = False)) == True]
```

```python
#rename all pnas variants to PNAS and add to new_data
data_new['new_j_title'] = np.where(data_original['j_title'].str.contains('NATIONAL ACADEMY OF SCIENCES',
                                                                         case = False, 
                                                                         regex = False) == True,
                                   'PNAS', '')
data_new['new_j_title'] = np.where(data_original['j_title'].str.contains('PNAS',
                                                                         case = False,
                                                                         regex = False) == True,
                                  'PNAS', '')                                  


#could these conditions be combined into one statement?
```

```python
#analyze journal title with word 'PARASIT'
data_original[data_original['j_title'].str.contains('PARASIT',
                                                    case = False,
                                                    regex = False,
                                                    na = False)]
```
