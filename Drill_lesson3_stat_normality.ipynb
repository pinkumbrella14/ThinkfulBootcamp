```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### 1. Exponential Disctribution

```python
#generate exponential distribution
exp = np.random.exponential(0.5, 100)

plt.hist(exp, color = 'green')

plt.show()
```

```python
plt.hist(exp, color = 'green')

#add vertical line at the mean
plt.axvline(exp.mean(), color = 'pink', linestyle = 'solid', linewidth = 2)

#add vertical lines at one std dev above mean
plt.axvline(exp.mean() + exp.std(), 
           color = 'pink',
           linestyle = 'dashed',
           linewidth = 2)

#add vertical line at one std dev below mean
plt.axvline(exp.mean() - exp.std(),
           color = 'pink',
           linestyle = 'dashed',
           linewidth = 2)

plt.show()
```

The mean and standard deviation of an exponential distribution are equal, so the standard deviation is not a particularly useful measure. However, the mean itself is fairly useful, as long as we keep in mind the fact that the disctribution is skewed, so the mean will not represent the 'center' of the distribution, but rather it's center of mass.

### 2. Ligistic distribution

```python
#generate logistic distribution
#location or mean, median, number)
logistic = np.random.logistic(0.5, 2.5, 100)

plt.hist(logistic, color = 'blue')

plt.show()
```

```python
plt.hist(logistic, color = 'blue')

plt.axvline(logistic.mean(),
            color = 'pink',
            linestyle = 'solid',
            linewidth = 2)

#plt.axvline(logistic.median(),
 #           color = 'red',
  #          linestyle = 'solid',
   #         linewidth = 0.5)

plt.axvline(logistic.mean() + logistic.std(),
            color = 'pink',
            linestyle = 'dashed',
            linewidth = 2)

plt.axvline(logistic.mean() - logistic.std(),
            color = 'pink',
            linestyle = 'dashed',
            linewidth = 2)

plt.show()
```

Here the mean and standard deviation are pretty useful, because this is a relatively symmertical distribution and in this case was modeled with mean and median fairly close to each other.

### 3. Uniform distribution

```python
uni = np.random.uniform(0, 11, 100)

plt.hist(uni, color = 'indigo')

plt.show()
```

```python
plt.hist(uni, color = 'indigo')

plt.axvline(uni.mean(),
            color = 'pink',
            linestyle = 'solid',
            linewidth = 2)

plt.axvline(uni.mean() + uni.std(),
            color = 'pink',
            linestyle = 'dashed',
            linewidth = 2)

plt.axvline(uni.mean() - uni.std(),
            color = 'pink',
            linestyle = 'dashed',
            linewidth = 2)

plt.show()
```

The uniform distribution equally distributes the numbers between the endpoints specified, so the mean id a true measure of tendency and the standard deviation confirms the equal weights of the numbers distributed.

### 4. Triangular distribution

```python
tri = np.random.triangular(1, 2, 11, 100)

plt.hist(tri, color = 'brown')

plt.show()
```

```python
plt.hist(tri, color = 'brown')

plt.axvline(tri.mean(),
            color = 'pink',
            linestyle = 'solid',
            linewidth = 2)

plt.axvline(tri.mean() + tri.std(),
            color = 'pink',
            linestyle = 'dashed',
            linewidth = 2)

plt.axvline(tri.mean() - tri.std(),
            color = 'pink',
            linestyle = 'dashed',
            linewidth = 2)

plt.show()
```

The mean and standard deviation are not particularly useful measures here, because the most numbers in the distribution are not near where the mean is and the distribution of values around the mean does not seem uniform.

### 5. Weibull distribution

```python
wei = np.random.weibull(15, 100)

plt.hist(wei, color = 'darkslategray')

plt.show()
```

```python
plt.hist(wei, color = 'darkslategray')

plt.axvline(wei.mean(),
            color = 'pink',
            linestyle = 'solid',
            linewidth = 2)

plt.axvline(wei.mean() + wei.std(),
            color = 'pink',
            linestyle = 'dashed',
            linewidth = 2)

plt.axvline(wei.mean() - wei.std(),
            color = 'pink',
            linestyle = 'dashed',
            linewidth = 2)

plt.show()
```

The mean does not seem a very accurate represenation of the central tendency of this distributionand the amount of values in the interval from the mean to each side of the distribution is not really represented well with the standard deviation.

### 6. t distribution

```python
t = np.random.standard_t(2, 100)

plt.hist(t, color = 'burlywood')

plt.show()
```

```python
plt.hist(t, color = 'burlywood')

plt.axvline(t.mean(),
            color = 'pink',
            linestyle = 'solid',
            linewidth = 2)

plt.axvline(t.mean() + t.std(),
            color = 'pink',
            linestyle = 'dashed',
            linewidth = 2)

plt.axvline(t.mean() - t.std(),
            color = 'pink',
            linestyle = 'dashed',
            linewidth = 2)

plt.show()
```

The t distribution is very similar to normal, so the mean and standard deviation are really good statistical measures.

### Working with 2 normally distributed variables

```python
df = pd.DataFrame()
df['rand1'] = np.random.normal(5, 0.5, 1000)
df['rand2'] = np.random.normal(10, 1, 1000)
df['sum'] = df['rand1'] + df['rand2']
```

```python
plt.hist(df['sum'], color = 'coral')

plt.show()
```

```python
plt.hist(df['sum'], color = 'coral')

plt.axvline(df['sum'].mean(),
            color = 'pink',
            linestyle = 'solid',
            linewidth = 2)

plt.axvline(df['sum'].mean() + df['sum'].std(),
            color = 'pink',
            linestyle = 'dashed',
            linewidth = 2)

plt.axvline(df['sum'].mean() - df['sum'].std(),
            color = 'pink',
            linestyle = 'dashed',
            linewidth = 2)

plt.show()
```

It seems like the sum of two normally distributed random variables is also normally distributed.
